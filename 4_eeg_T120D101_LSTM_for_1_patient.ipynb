# Divide into train/test data

```python
df_no_med_train_test = ['train'] * int(0.8 * len(df_no_med)) + ['test'] * (len(df_no_med) - int(0.8 * len(df_no_med)))
df_no_med['train_test'] = df_no_med_train_test
```

```python
df_with_med_train_test = ['train'] * int(0.8 * len(df_with_med)) + ['test'] * (len(df_with_med) - int(0.8 * len(df_with_med)))
df_with_med['train_test'] = df_with_med_train_test
```

```python
# Concatenate the dataframes
df = pd.concat([df_no_med, df_with_med])
```

```python
df.head()
```

```python
df_train = df[df['train_test'] == "train"]
df_test = df[df['train_test'] == "test"]
```

```python
df_train = df_train.drop('train_test', axis=1)
df_test = df_test.drop('train_test', axis=1)
```

# Features preprocessing

```python
features = list(df_train.columns)
features.remove("med")
```

```python
features
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

scaler = scaler.fit(df_train[features])

df_train.loc[:, features] = scaler.transform(df_train[features].to_numpy())
df_test.loc[:, features] = scaler.transform(df_test[features].to_numpy())
```

# Dataset preparation for classification

```python
# Sampling rate (frequency) is 1024 Hz (1024 samples in 1 second)
# 5120 samples in 5 seconds
```

```python
from scipy import stats

def create_dataset(X, y, time_steps=1, step=1):
    Xs, ys = [], []
    for i in range(0, len(X) - time_steps, step):
        v = X.iloc[i:(i + time_steps)].values
        labels = y.iloc[i: i + time_steps]
        Xs.append(v)        
        ys.append(pd.Series(labels).mode()[0])  # Use pandas' mode function
    return np.array(Xs), np.array(ys).reshape(-1, 1)

TIME_STEPS = 5120
STEP = 1024

X_train, y_train = create_dataset(
    df_train[features], 
    df_train.med, 
    TIME_STEPS, 
    STEP
)

X_test, y_test = create_dataset(
    df_test[features], 
    df_test.med, 
    TIME_STEPS, 
    STEP
)
```

```python
print(X_train.shape, y_train.shape)
```

```python
# Import label encoder 
from sklearn import preprocessing 
  
# label_encoder object knows  
# how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 

label_encoder = label_encoder.fit(y_train)

y_train = label_encoder.transform(y_train)
y_test = label_encoder.transform(y_test)
```

```python
y_train = y_train.reshape(-1, 1) 
y_test = y_test.reshape(-1, 1) 
```

```python
print(X_train.shape, y_train.shape)
```

# Classification with LSTM

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import seaborn as sns
from pylab import rcParams
import matplotlib.pyplot as plt
from matplotlib import rc
from pandas.plotting import register_matplotlib_converters
```

```python
model = keras.Sequential()
model.add(
    keras.layers.Bidirectional(
      keras.layers.LSTM(
          units=128, 
          input_shape=[X_train.shape[1], X_train.shape[2]]
      )
    )
)
model.add(keras.layers.Dropout(rate=0.5))
model.add(keras.layers.Dense(units=128, activation='relu'))
model.add(keras.layers.Dense(y_train.shape[1], activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
```

```python
history = model.fit(
    X_train, y_train,
    epochs=5,
    batch_size=32,
    validation_split=0.1,
    shuffle=False
)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```
