```python
import numpy as np
import mne
import pandas as pd

# Import the EEGExtract library
# https://github.com/sari-saba-sadiya/EEGExtract/tree/main
#import EEGExtract as eeg
##############################################

# Initialize dictionaries to store data for all patients
data_no_med = {}
data_with_med = {}

# Assume patient_ids is a list of identifiers for the patients
patient_ids = ['patient_2024_04_24', 'patient_2024_05_07', 'patient_2024_02_21'] 

fs = 1024

# Loop through each patient and read their data
for patient_id in patient_ids:
    # Read the raw data for each state and patient
    raw_no_med = mne.io.read_raw_fif(f"{patient_id}_raw_no_med.fif", preload=True)
    raw_with_med = mne.io.read_raw_fif(f"{patient_id}_raw_with_med.fif", preload=True)

    #Check information
    display(raw_no_med.info)
    display(raw_with_med.info)

    #Plot the data
    raw_no_med.plot();
    raw_with_med.plot();

    # Create fixed length epochs for each state and patient
    epochs_no_med = mne.make_fixed_length_epochs(raw_no_med, duration=5.0, overlap=0.0, preload=True)
    epochs_with_med = mne.make_fixed_length_epochs(raw_with_med, duration=5.0, overlap=0.0, preload=True)

    # Get the data and transpose it to the required shape 3D np array [chans x ms x epochs] 
    epochs_no_med_data_transposed = epochs_no_med.get_data().transpose(1, 2, 0)
    epochs_with_med_data_transposed = epochs_with_med.get_data().transpose(1, 2, 0)

    # Store the transposed data in the dictionaries
    data_no_med[patient_id] = epochs_no_med_data_transposed
    data_with_med[patient_id] = epochs_with_med_data_transposed

# Concatenate the data for each state across all patients
all_data_no_med = np.concatenate(list(data_no_med.values()), axis=2)
all_data_with_med = np.concatenate(list(data_with_med.values()), axis=2)

##############################################
# Features extraction
# Lyapunov Exponent
#separation between signals with similar trajectories
LyapunovRes_no_med = eeg.lyapunov(all_data_no_med)
LyapunovRes_with_med = eeg.lyapunov(all_data_with_med)

# Fractal Embedding Dimension
#how signal properties change with scale

HiguchiFD_Res_no_med  = eeg.hFD(all_data_no_med,3)
HiguchiFD_Res_with_med  = eeg.hFD(all_data_with_med,3)

#HiguchiFD_Res_no_med  = eeg.hFD(all_data_no_med[0,:,0],3)
#HiguchiFD_Res_with_med  = eeg.hFD(all_data_with_med[0,:,0],3)

# Median Frequency
#the median spectral power
medianFreqRes_no_med  = eeg.medianFreq(all_data_no_med,fs)
medianFreqRes_with_med  = eeg.medianFreq(all_data_with_med,fs)

# α band Power
#spectral power in the 8-12Hz range
bandPwr_alpha_no_med = eeg.bandPower(all_data_no_med, 8, 12, fs)
bandPwr_alpha_with_med = eeg.bandPower(all_data_with_med, 8, 12, fs)

# β band Power
#spectral power in the 12-30Hz range
bandPwr_beta_no_med = eeg.bandPower(all_data_no_med, 12, 30, fs)
bandPwr_beta_with_med = eeg.bandPower(all_data_with_med, 12, 30, fs)

# Hjorth Parameters
# Calculate Hjorth Mobility and Complexity
Hjorth_no_med = eeg.hjorthParameters(all_data_no_med)
Hjorth_with_med = eeg.hjorthParameters(all_data_with_med)

# The hjorthParameters function returns a tuple with (mobility, complexity)
# separate parameters like this:
mobility_no_med, complexity_no_med = Hjorth_no_med
mobility_with_med, complexity_with_med = Hjorth_with_med

##############################################

from scipy import stats

# Function to perform Mann-Whitney U test for a given feature
def perform_mannwhitney_test(feature_no_med, feature_with_med):
    u_values, p_values = [], []
    for i in range(feature_no_med.shape[0]):  # iterate over channels
        u_stat, p_val = stats.mannwhitneyu(feature_no_med[i], feature_with_med[i])
        u_values.append(u_stat)
        p_values.append(p_val)
    return u_values, p_values

# Perform the test for each feature and create a DataFrame for the results
def create_stat_results_df(feature_name, feature_no_med, feature_with_med, ch_names):
    u_values, p_values = perform_mannwhitney_test(feature_no_med, feature_with_med)
    df = pd.DataFrame({
        'Channel': ch_names,
        f'{feature_name} U-Value': u_values,
        f'{feature_name} P-Value': p_values
    })
    return df


# Create DataFrames for each feature
df_lyapunov = create_stat_results_df('Lyapunov', LyapunovRes_no_med, LyapunovRes_with_med, raw_no_med.info['ch_names'])
#df_higuchi = create_stat_results_df('HiguchiFD', HiguchiFD_Res_no_med, HiguchiFD_Res_with_med, raw_no_med.info['ch_names'])
df_median_freq = create_stat_results_df('MedianFreq', medianFreqRes_no_med, medianFreqRes_with_med, raw_no_med.info['ch_names'])
df_band_power_alpha = create_stat_results_df('BandPowerAlpha', bandPwr_alpha_no_med, bandPwr_alpha_with_med, raw_no_med.info['ch_names'])
df_band_power_beta = create_stat_results_df('BandPowerBeta', bandPwr_beta_no_med, bandPwr_beta_with_med, raw_no_med.info['ch_names'])
df_hjorth_mobility = create_stat_results_df('HjorthMobility', mobility_no_med, mobility_with_med, raw_no_med.info['ch_names'])
df_hjorth_complexity = create_stat_results_df('HjorthComplexity', complexity_no_med, complexity_with_med, raw_no_med.info['ch_names'])


# Combine all DataFrames into one
df_combined = df_lyapunov.join([#df_higuchi.drop('Channel', axis=1), 
                                df_median_freq.drop('Channel', axis=1), 
                                df_band_power_beta.drop('Channel', axis=1),
                                df_band_power_alpha.drop('Channel', axis=1),
                                df_hjorth_mobility.drop('Channel', axis=1), 
                                df_hjorth_complexity.drop('Channel', axis=1)
                                ])

# Display the combined DataFrame
display(df_combined)

##############################################
##############################################

# Filter the DataFrame to include only rows where the P-value is less than 0.05
significant_channels_df_lyapunov = df_lyapunov[df_lyapunov['Lyapunov P-Value'] < 0.05]
#significant_channels_df_higuchi = df_higuchi[df_higuchi['HiguchiFD P-Value'] < 0.05]
significant_channels_df_median_freq = df_median_freq[df_median_freq['MedianFreq P-Value'] < 0.05]
significant_channels_df_band_power_alpha = df_band_power_alpha[df_band_power_alpha['BandPowerAlpha P-Value'] < 0.05]
significant_channels_df_band_power_beta = df_band_power_beta[df_band_power_beta['BandPowerBeta P-Value'] < 0.05]
significant_channels_df_hjorth_mobility = df_hjorth_mobility[df_hjorth_mobility['HjorthMobility P-Value'] < 0.05]
significant_channels_df_hjorth_complexity = df_hjorth_complexity[df_hjorth_complexity['HjorthComplexity P-Value'] < 0.05]

# Display the filtered DataFrame
#display(significant_channels_df_lyapunov)
print(significant_channels_df_lyapunov.shape)

#display(significant_channels_df_higuchi)
#print(significant_channels_df_higuchi.shape)

#display(significant_channels_df_median_freq)
print(significant_channels_df_median_freq.shape)

#display(significant_channels_df_band_power_alpha)
print(significant_channels_df_band_power_alpha.shape)

#display(significant_channels_df_band_power_beta)
print(significant_channels_df_band_power_beta.shape)

#display(significant_channels_df_hjorth_mobility)
print(significant_channels_df_hjorth_mobility.shape)

#display(significant_channels_df_hjorth_complexity)
print(significant_channels_df_hjorth_complexity.shape)

#significant_channels_df_lyapunov_list = significant_channels_df_lyapunov['Channel'].tolist()
#significant_channels_df_higuchi_list = significant_channels_df_higuchi['Channel'].tolist()
#significant_channels_df_median_freq_list = significant_channels_df_median_freq['Channel'].tolist()
#significant_channels_df_band_power_alpha_list = significant_channels_df_band_power_alpha['Channel'].tolist()
#significant_channels_df_band_power_beta_list = significant_channels_df_band_power_beta['Channel'].tolist()
#significant_channels_df_hjorth_mobility_list = significant_channels_df_hjorth_mobility['Channel'].tolist()
#significant_channels_df_hjorth_complexity_list = significant_channels_df_hjorth_complexity['Channel'].tolist()

##############################################

import matplotlib.pyplot as plt

# Function to create box plots for statistically significant channels
def create_significant_box_plots(feature_name, feature_results_no_med, feature_results_with_med, ch_names, p_values):
    for i, ch_name in enumerate(ch_names):
        # Check if the p-value is less than 0.05 for statistical significance
        if p_values[i] < 0.05:
            # Data for the 'No Medication' and 'With Medication' conditions
            data = [feature_results_no_med[i, :], feature_results_with_med[i, :]]
            
            # Create a figure for the current channel
            fig, ax = plt.subplots(figsize=(5, 4))
            
            # Create a boxplot for the current channel
            ax.boxplot(data, positions=[1, 2], widths=0.6, notch=True, patch_artist=True, whis=2.5)
            
            # Set the title and labels
            ax.set_title(f'Significant {feature_name} - Channel: {ch_name}')
            ax.set_xticks([1, 2])
            ax.set_xticklabels(['No Med', 'With Med'])
            
            # Save the plot with a unique name for each channel
            #plt.savefig(f'significant_{feature_name}_{ch_name}_boxplot.png')
            #plt.close(fig)  # Close the figure to free up memory

# Call the function for Lyapunov feature
#create_significant_box_plots('Lyapunov', LyapunovRes_no_med, LyapunovRes_with_med, raw_no_med.info['ch_names'], df_lyapunov['Lyapunov P-Value'].values)

# Uncomment and call the function for HiguchiFD feature when ready
#create_significant_box_plots('HiguchiFD', HiguchiFD_Res_no_med, HiguchiFD_Res_with_med, raw_no_med.info['ch_names'], df_higuchi['HiguchiFD P-Value'].values)

# Call the function for MedianFreq feature
create_significant_box_plots('MedianFreq', medianFreqRes_no_med, medianFreqRes_with_med, raw_no_med.info['ch_names'], df_median_freq['MedianFreq P-Value'].values)

# Call the function for BandPowerAlpha feature
#create_significant_box_plots('BandPowerAlpha', bandPwr_alpha_no_med, bandPwr_alpha_with_med, raw_no_med.info['ch_names'], df_band_power_alpha['BandPoweralpha P-Value'].values)

# Call the function for BandPowerBeta feature
#create_significant_box_plots('BandPowerBeta', bandPwr_beta_no_med, bandPwr_beta_with_med, raw_no_med.info['ch_names'], df_band_power_beta['BandPowerBeta P-Value'].values)

# Call the function for HjorthMobility feature
#create_significant_box_plots('HjorthMobility', mobility_no_med, mobility_with_med, raw_no_med.info['ch_names'], df_hjorth_mobility['HjorthMobility P-Value'].values)

# Call the function for HjorthComplexity feature
#create_significant_box_plots('HjorthComplexity', complexity_no_med, complexity_with_med, raw_no_med.info['ch_names'], df_hjorth_complexity['HjorthComplexity P-Value'].values)
```

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Function to create a simple neural network model with sigmoid activation
def create_model(input_shape):
    model = Sequential([
        Dense(64, activation='relu', input_shape=(input_shape,)),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Function to perform cross-validation and evaluate different models
def evaluate_models(X, y, feature_name):
    # Define the cross-validation parameters
    n_splits = 5
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    
    # Initialize the scores
    scores_nn_acc = []
    scores_nn_auc = []
    scores_rf_acc = []
    scores_rf_auc = []
    scores_lr_acc = []
    scores_lr_auc = []
    scores_svm_acc = []
    scores_svm_auc = []
    scores_xgb_acc = []
    scores_xgb_auc = []
    
    # Loop over each fold
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        # Scale the features
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)
        
        # Neural Network
        nn_model = create_model(X_train_scaled.shape[1])
        nn_model.fit(X_train_scaled, y_train, epochs=10, batch_size=32, verbose=0)
        nn_predictions = nn_model.predict(X_test_scaled).flatten()
        scores_nn_acc.append(accuracy_score(y_test, (nn_predictions > 0.5).astype(int)))
        scores_nn_auc.append(roc_auc_score(y_test, nn_predictions))
        
        # Random Forest
        rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
        rf_model.fit(X_train_scaled, y_train)
        rf_predictions = rf_model.predict(X_test_scaled)
        scores_rf_acc.append(accuracy_score(y_test, rf_predictions))
        scores_rf_auc.append(roc_auc_score(y_test, rf_model.predict_proba(X_test_scaled)[:, 1]))
        
        # Logistic Regression
        lr_model = LogisticRegression(random_state=42)
        lr_model.fit(X_train_scaled, y_train)
        lr_predictions = lr_model.predict(X_test_scaled)
        scores_lr_acc.append(accuracy_score(y_test, lr_predictions))
        scores_lr_auc.append(roc_auc_score(y_test, lr_model.predict_proba(X_test_scaled)[:, 1]))
        
        # Support Vector Machine
        svm_model = SVC(probability=True, random_state=42)
        svm_model.fit(X_train_scaled, y_train)
        svm_predictions = svm_model.predict(X_test_scaled)
        scores_svm_acc.append(accuracy_score(y_test, svm_predictions))
        scores_svm_auc.append(roc_auc_score(y_test, svm_model.predict_proba(X_test_scaled)[:, 1]))
        
        # XGBoost
        xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
        xgb_model.fit(X_train_scaled, y_train)
        xgb_predictions = xgb_model.predict(X_test_scaled)
        scores_xgb_acc.append(accuracy_score(y_test, xgb_predictions))
        scores_xgb_auc.append(roc_auc_score(y_test, xgb_model.predict_proba(X_test_scaled)[:, 1]))
    
    # Calculate the average score across all folds
    avg_score_nn_acc = round(np.mean(scores_nn_acc), 3)
    avg_score_nn_auc = round(np.mean(scores_nn_auc), 3)
    avg_score_rf_acc = round(np.mean(scores_rf_acc), 3)
    avg_score_rf_auc = round(np.mean(scores_rf_auc), 3)
    avg_score_lr_acc = round(np.mean(scores_lr_acc), 3)
    avg_score_lr_auc = round(np.mean(scores_lr_auc), 3)
    avg_score_svm_acc = round(np.mean(scores_svm_acc), 3)
    avg_score_svm_auc = round(np.mean(scores_svm_auc), 3)
    avg_score_xgb_acc = round(np.mean(scores_xgb_acc), 3)
    avg_score_xgb_auc = round(np.mean(scores_xgb_auc), 3)
    
    # Print results
    print(f'Feature: {feature_name}')
    print(f'Neural Network - Average Accuracy: {avg_score_nn_acc}, Average AUC: {avg_score_nn_auc}')
    print(f'Random Forest - Average Accuracy: {avg_score_rf_acc}, Average AUC: {avg_score_rf_auc}')
    print(f'Logistic Regression - Average Accuracy: {avg_score_lr_acc}, Average AUC: {avg_score_lr_auc}')
    print(f'Support Vector Machine - Average Accuracy: {avg_score_svm_acc}, Average AUC: {avg_score_svm_auc}')
    print(f'XGBoost - Average Accuracy: {avg_score_xgb_acc}, Average AUC: {avg_score_xgb_auc}')

# Function to prepare the data and call evaluate_models
def evaluate_feature(feature_no_med, feature_with_med, feature_name):
    # Combine the features for 'no_med' and 'med'
    X = np.vstack((feature_no_med, feature_with_med))
    y = np.array([0] * len(feature_no_med) + [1] * len(feature_with_med))  # 0 for 'no_med', 1 for 'med'
    
    # Perform cross-validation
    evaluate_models(X, y, feature_name)

# Assuming the feature extraction code provided earlier is executed here

# Evaluate each feature set
evaluate_feature(medianFreqRes_no_med, medianFreqRes_with_med, 'Median Frequency')
evaluate_feature(bandPwr_beta_no_med, bandPwr_beta_with_med, 'Beta Band Power')
evaluate_feature(mobility_no_med, mobility_with_med, 'Hjorth Mobility')
evaluate_feature(complexity_no_med, complexity_with_med, 'Hjorth Complexity')
```

```python
import numpy as np
from scipy.stats import friedmanchisquare
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Function to create a simple neural network model with sigmoid activation
def create_model(input_shape):
    model = Sequential([
        Dense(64, activation='relu', input_shape=(input_shape,)),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Function to perform cross-validation and evaluate different models
def evaluate_models(X, y, feature_name):
    # Define the cross-validation parameters
    n_splits = 5
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    
    # Initialize the scores
    scores_nn_acc = []
    scores_rf_acc = []
    scores_lr_acc = []
    scores_svm_acc = []
    scores_xgb_acc = []
    
    # Loop over each fold
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        # Scale the features
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)
        
        # Neural Network
        nn_model = create_model(X_train_scaled.shape[1])
        nn_model.fit(X_train_scaled, y_train, epochs=10, batch_size=32, verbose=0)
        nn_predictions = nn_model.predict(X_test_scaled).flatten()
        scores_nn_acc.append(accuracy_score(y_test, (nn_predictions > 0.5).astype(int)))
        
        # Random Forest
        rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
        rf_model.fit(X_train_scaled, y_train)
        rf_predictions = rf_model.predict(X_test_scaled)
        scores_rf_acc.append(accuracy_score(y_test, rf_predictions))
        
        # Logistic Regression
        lr_model = LogisticRegression(random_state=42)
        lr_model.fit(X_train_scaled, y_train)
        lr_predictions = lr_model.predict(X_test_scaled)
        scores_lr_acc.append(accuracy_score(y_test, lr_predictions))
        
        # Support Vector Machine
        svm_model = SVC(probability=True, random_state=42)
        svm_model.fit(X_train_scaled, y_train)
        svm_predictions = svm_model.predict(X_test_scaled)
        scores_svm_acc.append(accuracy_score(y_test, svm_predictions))
        
        # XGBoost
        xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
        xgb_model.fit(X_train_scaled, y_train)
        xgb_predictions = xgb_model.predict(X_test_scaled)
        scores_xgb_acc.append(accuracy_score(y_test, xgb_predictions))
    
    # Perform Friedman tests between models for accuracy
    print(f'Performing Friedman tests between models for accuracy on feature: {feature_name}')
    statistic_acc, p_value_acc = friedmanchisquare(scores_nn_acc, scores_rf_acc, scores_lr_acc, scores_svm_acc, scores_xgb_acc)
    print(f'Friedman test for accuracy on feature {feature_name} - Test Statistic: {statistic_acc}, P-value: {p_value_acc}')
    
    # Return the scores for this feature
    return scores_nn_acc, scores_rf_acc, scores_lr_acc, scores_svm_acc, scores_xgb_acc

# Initialize dictionaries to store the accuracies for each feature
accuracies = {}

# Function to prepare the data and call evaluate_models
def evaluate_feature(feature_no_med, feature_with_med, feature_name):
    # Combine the features for 'no_med' and 'med'
    X = np.vstack((feature_no_med, feature_with_med))
    y = np.array([0] * len(feature_no_med) + [1] * len(feature_with_med))  # 0 for 'no_med', 1 for 'med'
    
    # Perform cross-validation
    acc_nn, acc_rf, acc_lr, acc_svm, acc_xgb = evaluate_models(X, y, feature_name)
    
    # Store the accuracies for this feature
    accuracies[feature_name] = [acc_nn, acc_rf, acc_lr, acc_svm, acc_xgb]

# Assuming the feature extraction code provided earlier is executed here

# Evaluate each feature set
evaluate_feature(medianFreqRes_no_med, medianFreqRes_with_med, 'Median Frequency')
evaluate_feature(bandPwr_beta_no_med, bandPwr_beta_with_med, 'Beta Band Power')
evaluate_feature(mobility_no_med, mobility_with_med, 'Hjorth Mobility')
evaluate_feature(complexity_no_med, complexity_with_med, 'Hjorth Complexity')

# Perform Friedman test for accuracy
print('Performing Friedman test for accuracy across features...')
# Prepare the accuracies for the Friedman test
all_accuracies = []
for feature_name in accuracies:
    for model_accuracies in accuracies[feature_name]:
        all_accuracies.append(model_accuracies)
# Perform the Friedman test
statistic_acc, p_value_acc = friedmanchisquare(*all_accuracies)
print(f'Friedman test for accuracy - Test Statistic: {statistic_acc}, P-value: {p_value_acc}')
```

```python
import numpy as np
from scipy.stats import friedmanchisquare
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Function to create a simple neural network model with sigmoid activation
def create_model(input_shape):
    model = Sequential([
        Dense(64, activation='relu', input_shape=(input_shape,)),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Function to perform cross-validation and evaluate different models
def evaluate_models(X, y, feature_name):
    # Define the cross-validation parameters
    n_splits = 5
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    
    # Initialize the scores
    scores_nn_acc = []
    scores_rf_acc = []
    scores_lr_acc = []
    scores_svm_acc = []
    scores_xgb_acc = []
    
    scores_nn_auc = []
    scores_rf_auc = []
    scores_lr_auc = []
    scores_svm_auc = []
    scores_xgb_auc = []
    
    # Loop over each fold
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        # Scale the features
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)
        
        # Neural Network
        nn_model = create_model(X_train_scaled.shape[1])
        nn_model.fit(X_train_scaled, y_train, epochs=10, batch_size=32, verbose=0)
        nn_predictions = nn_model.predict(X_test_scaled).flatten()
        scores_nn_acc.append(accuracy_score(y_test, (nn_predictions > 0.5).astype(int)))
        scores_nn_auc.append(roc_auc_score(y_test, nn_predictions))
        
        # Random Forest
        rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
        rf_model.fit(X_train_scaled, y_train)
        rf_predictions = rf_model.predict(X_test_scaled)
        rf_proba = rf_model.predict_proba(X_test_scaled)[:, 1]
        scores_rf_acc.append(accuracy_score(y_test, rf_predictions))
        scores_rf_auc.append(roc_auc_score(y_test, rf_proba))
        
        # Logistic Regression
        lr_model = LogisticRegression(random_state=42)
        lr_model.fit(X_train_scaled, y_train)
        lr_predictions = lr_model.predict(X_test_scaled)
        lr_proba = lr_model.predict_proba(X_test_scaled)[:, 1]
        scores_lr_acc.append(accuracy_score(y_test, lr_predictions))
        scores_lr_auc.append(roc_auc_score(y_test, lr_proba))
        
        # Support Vector Machine
        svm_model = SVC(probability=True, random_state=42)
        svm_model.fit(X_train_scaled, y_train)
        svm_predictions = svm_model.predict(X_test_scaled)
        svm_proba = svm_model.predict_proba(X_test_scaled)[:, 1]
        scores_svm_acc.append(accuracy_score(y_test, svm_predictions))
        scores_svm_auc.append(roc_auc_score(y_test, svm_proba))
        
        # XGBoost
        xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
        xgb_model.fit(X_train_scaled, y_train)
        xgb_predictions = xgb_model.predict(X_test_scaled)
        xgb_proba = xgb_model.predict_proba(X_test_scaled)[:, 1]
        scores_xgb_acc.append(accuracy_score(y_test, xgb_predictions))
        scores_xgb_auc.append(roc_auc_score(y_test, xgb_proba))
    
    # Print mean and standard deviation for accuracy and AUC
    print(f'Accuracy scores for feature: {feature_name}')
    print(f'Neural Network: Mean = {np.mean(scores_nn_acc):.4f}, Std = {np.std(scores_nn_acc):.4f}')
    print(f'Random Forest: Mean = {np.mean(scores_rf_acc):.4f}, Std = {np.std(scores_rf_acc):.4f}')
    print(f'Logistic Regression: Mean = {np.mean(scores_lr_acc):.4f}, Std = {np.std(scores_lr_acc):.4f}')
    print(f'SVM: Mean = {np.mean(scores_svm_acc):.4f}, Std = {np.std(scores_svm_acc):.4f}')
    print(f'XGBoost: Mean = {np.mean(scores_xgb_acc):.4f}, Std = {np.std(scores_xgb_acc):.4f}')
    
    print(f'AUC scores for feature: {feature_name}')
    print(f'Neural Network: Mean = {np.mean(scores_nn_auc):.4f}, Std = {np.std(scores_nn_auc):.4f}')
    print(f'Random Forest: Mean = {np.mean(scores_rf_auc):.4f}, Std = {np.std(scores_rf_auc):.4f}')
    print(f'Logistic Regression: Mean = {np.mean(scores_lr_auc):.4f}, Std = {np.std(scores_lr_auc):.4f}')
    print(f'SVM: Mean = {np.mean(scores_svm_auc):.4f}, Std = {np.std(scores_svm_auc):.4f}')
    print(f'XGBoost: Mean = {np.mean(scores_xgb_auc):.4f}, Std = {np.std(scores_xgb_auc):.4f}')
    
    # Perform Friedman tests between models for accuracy and AUC
    print(f'Performing Friedman tests between models for feature: {feature_name}')
    statistic_acc, p_value_acc = friedmanchisquare(scores_nn_acc, scores_rf_acc, scores_lr_acc, scores_svm_acc, scores_xgb_acc)
    statistic_auc, p_value_auc = friedmanchisquare(scores_nn_auc, scores_rf_auc, scores_lr_auc, scores_svm_auc, scores_xgb_auc)
    print(f'Friedman test for accuracy on feature {feature_name} - Test Statistic: {statistic_acc:.4f}, P-value: {p_value_acc:.4f}')
    print(f'Friedman test for AUC on feature {feature_name} - Test Statistic: {statistic_auc:.4f}, P-value: {p_value_auc:.4f}')
    
    # Return the scores for this feature
    return (scores_nn_acc, scores_rf_acc, scores_lr_acc, scores_svm_acc, scores_xgb_acc), \
           (scores_nn_auc, scores_rf_auc, scores_lr_auc, scores_svm_auc, scores_xgb_auc)

# Initialize dictionaries to store the accuracies and AUCs for each feature
accuracies = {}
aucs = {}

# Function to prepare the data and call evaluate_models
def evaluate_feature(feature_no_med, feature_with_med, feature_name):
    # Combine the features for 'no_med' and 'med'
    X = np.vstack((feature_no_med, feature_with_med))
    y = np.array([0] * len(feature_no_med) + [1] * len(feature_with_med))  # 0 for 'no_med', 1 for 'med'
    
    # Perform cross-validation
    acc_scores, auc_scores = evaluate_models(X, y, feature_name)
    
    # Store the accuracies and AUCs for this feature
    accuracies[feature_name] = acc_scores
    aucs[feature_name] = auc_scores

# Assuming the feature extraction code provided earlier is executed here

# Evaluate each feature set
evaluate_feature(medianFreqRes_no_med, medianFreqRes_with_med, 'Median Frequency')
evaluate_feature(bandPwr_beta_no_med, bandPwr_beta_with_med, 'Beta Band Power')
evaluate_feature(mobility_no_med, mobility_with_med, 'Hjorth Mobility')
evaluate_feature(complexity_no_med, complexity_with_med, 'Hjorth Complexity')

# Perform Friedman test for accuracy and AUC across features
print('Performing Friedman test for accuracy across features...')
all_accuracies = []
for feature_name in accuracies:
    for model_accuracies in accuracies[feature_name]:
        all_accuracies.append(model_accuracies)

statistic_acc, p_value_acc = friedmanchisquare(*all_accuracies)
print(f'Friedman test for accuracy - Test Statistic: {statistic_acc:.4f}, P-value: {p_value_acc:.4f}')

print('Performing Friedman test for AUC across features...')
all_aucs = []
for feature_name in aucs:
    for model_aucs in aucs[feature_name]:
        all_aucs.append(model_aucs)

statistic_auc, p_value_auc = friedmanchisquare(*all_aucs)
print(f'Friedman test for AUC - Test Statistic: {statistic_auc:.4f}, P-value: {p_value_auc:.4f}')
```

```python
import scikit_posthocs as sp

# Function to identify the best model for each feature based on Friedman test results and perform Nemenyi post-hoc test
def find_best_models_and_feature():
    nemenyi_results_acc = {}
    nemenyi_results_auc = {}
    best_models_acc = {}
    best_models_auc = {}

    models = ['Neural Network', 'Random Forest', 'Logistic Regression', 'SVM', 'XGBoost']

    for feature_name in accuracies:
        acc_data = np.array(accuracies[feature_name])
        auc_data = np.array(aucs[feature_name])

        statistic_acc, p_value_acc = friedmanchisquare(*acc_data)
        statistic_auc, p_value_auc = friedmanchisquare(*auc_data)

        print(f'Friedman test for accuracy on feature {feature_name} - Test Statistic: {statistic_acc:.4f}, P-value: {p_value_acc:.4f}')
        print(f'Friedman test for AUC on feature {feature_name} - Test Statistic: {statistic_auc:.4f}, P-value: {p_value_auc:.4f}')

        if p_value_acc < 0.05:
            nemenyi_acc = sp.posthoc_nemenyi_friedman(acc_data.T)
            nemenyi_results_acc[feature_name] = nemenyi_acc
            print(f'Nemenyi post-hoc test results for accuracy on feature {feature_name}:\n{nemenyi_acc}')
        else:
            nemenyi_results_acc[feature_name] = 'No significant difference (accuracy)'
            print(f'No significant difference in accuracy for feature {feature_name}')

        if p_value_auc < 0.05:
            nemenyi_auc = sp.posthoc_nemenyi_friedman(auc_data.T)
            nemenyi_results_auc[feature_name] = nemenyi_auc
            print(f'Nemenyi post-hoc test results for AUC on feature {feature_name}:\n{nemenyi_auc}')
        else:
            nemenyi_results_auc[feature_name] = 'No significant difference (AUC)'
            print(f'No significant difference in AUC for feature {feature_name}')

        best_model_index_acc = np.argmax([np.mean(acc) for acc in acc_data])
        best_model_index_auc = np.argmax([np.mean(auc) for auc in auc_data])

        best_models_acc[feature_name] = models[best_model_index_acc]
        best_models_auc[feature_name] = models[best_model_index_auc]

        # Print best model and its mean and std
        print(f'Best ML model for {feature_name} based on accuracy: {models[best_model_index_acc]}')
        print(f'Mean accuracy: {np.mean(acc_data[best_model_index_acc]):.4f}, Std: {np.std(acc_data[best_model_index_acc]):.4f}')
        print(f'Best ML model for {feature_name} based on AUC: {models[best_model_index_auc]}')
        print(f'Mean AUC: {np.mean(auc_data[best_model_index_auc]):.4f}, Std: {np.std(auc_data[best_model_index_auc]):.4f}')
        print()

    # Determine the best overall feature extraction method
    mean_accuracies = {feature_name: np.mean([np.mean(acc) for acc in accuracies[feature_name]]) for feature_name in accuracies}
    mean_aucs = {feature_name: np.mean([np.mean(auc) for auc in aucs[feature_name]]) for feature_name in aucs}

    best_feature_acc = max(mean_accuracies, key=mean_accuracies.get)
    best_feature_auc = max(mean_aucs, key=mean_aucs.get)

    # Print best overall feature extraction method and its mean and std
    print(f'Best feature extraction method based on accuracy: {best_feature_acc}')
    print(f'Mean accuracy: {np.mean([np.mean(acc) for acc in accuracies[best_feature_acc]]):.4f}, Std: {np.std([np.mean(acc) for acc in accuracies[best_feature_acc]]):.4f}')
    print(f'Best feature extraction method based on AUC: {best_feature_auc}')
    print(f'Mean AUC: {np.mean([np.mean(auc) for auc in aucs[best_feature_auc]]):.4f}, Std: {np.std([np.mean(auc) for auc in aucs[best_feature_auc]]):.4f}')

    return best_models_acc, best_models_auc, best_feature_acc, best_feature_auc, nemenyi_results_acc, nemenyi_results_auc

# Find best models and feature extraction method
best_models_acc, best_models_auc, best_feature_acc, best_feature_auc, nemenyi_results_acc, nemenyi_results_auc = find_best_models_and_feature()
```
